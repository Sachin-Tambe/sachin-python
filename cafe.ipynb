```python
total_bill = 0 
import time
def welcome_user():
    print("Welcome to the King Of Cafe !")
def show_menu():
    global total_bill 
    menu = {
        1 :("Coffie", 300),
        2 :('Tea', 200),
        3 :('Milk', 100),
        4 :('Juice', 150),
        5 :('Burger', 250),
        6 :('Pizza', 350),
        7 :('Sandwich', 200),
        8 :('Pasta', 300),
        9 :('Noodles', 250),
        10 :('French Fries', 150),
        11 :('Chicken', 250),
        12 :('Fish', 300),
        13 :('Rice', 150),
        14 :('Biryani', 250)
    }
    print("Menu")
    for key , value in menu.items():
        item_price , item_value = value
        print(f"{key}:{item_price} - {item_value}")
    chooise = int(input("Please Enter Your Choise As Per Dish Number "))
    if chooise in menu.keys():
        global total_price
        item_price , item_value = menu[chooise]
        print(f"Good Choise sir , You choice := {menu[chooise]}")
        quantity = int(input("The No Of Quantity You Want Sir ?"))
        print("Your Order Is Preparing ")
        time.sleep(5)
        total_price = item_value * quantity
        total_bill += total_price 
        user_want = input("Sir , You Want Anything else  Y/N ?")
        while user_want == 'Y'.lower() : 
            show_menu()
        else:
            get_payement()
    else:
        print("I Am So Soory Sir , This Item is Not Available In Our Cafe")
        show_menu()
def get_payement():
        global total_bill
        print(f"Sir Your Total Bill is {total_bill}")
        get_pay = int(input("Sir Please Enter Your Amount You Want To Pay"))
        if get_pay >= total_bill :
             return_amount = get_pay - total_bill
             print(f"Please Sir Take Your Return :=  {return_amount}")
        else :
            print("Payement Failed")
            remaining_amount =  total_bill - get_pay
            total_bill = remaining_amount 
            get_payement()
                
        
    
```
